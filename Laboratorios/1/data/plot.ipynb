```python
import pandas as pd
import numpy as np

raio_fixo = pd.read_csv('raio-fixo.csv')
angulo_fixo = pd.read_csv('angulo-fixo.csv')
lados_fixo = pd.read_csv('lados-fixo.csv')
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```
